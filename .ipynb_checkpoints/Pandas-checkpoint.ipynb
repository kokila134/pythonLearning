```python
#Pandas Series
import pandas as pd
s=pd.Series()        #Create empty series
print(s)
```

```python
#Pandas series from array
import pandas as pd
import numpy as np
ar=['a','b','c','d']    #array
s=pd.Series(ar)
print(s)
np_ar=np.array(ar)     #numpy array
s=pd.Series(np_ar)
print(s)

#passing index values
s=pd.Series(np_ar,index=[101,102,103,104])       #custom index values
print(s)

```

```python
#Pandas series from dict
import pandas as pd
dict={'a':[1,0.1],
     'b':[2,0.2],
     'c':[3,0.3],
     'd':[4,0.4]}
s=pd.Series(dict)      #takes key as index by default
print(s)

#passing index values
s=pd.Series(dict,index=['b','c','a','e','d'])
print(s)
```

```python
#Pandas series from scalar value
scalar=5
s=pd.Series(scalar,[0,1,2,3,4,5,6,7])      #need to provide index in order to make series from scalar value
print(s)
```

```python
#accessing elements of a series
import pandas as pd
s=pd.Series([1,2,3,4],index=['a','b','c','d'])
print(s[0])       #prints element at 0th position
print(s[:3])     #prints first 3 elements
print(s[-2:])     #prints last 2 elements
print(s[1:3])     #prints element at 1st and 2nd position
print(s[['b','c']] )   #prints elements using index values
```

```python
#DataFrame from List of Dicts
import pandas as pd
dict=[{'Name':'Aa','Age':10},
      {'Name':'Bb','Age':20,'Gender':'Male'}]
data=pd.DataFrame(dict)
print(data)

data=pd.DataFrame(dict,index=['First','Second'])       #assigning index values
print(data)

data=pd.DataFrame(dict,index=['First','Second'],columns=['Name','Age'])      #row and column indices
print(data)

data=pd.DataFrame(dict,index=['First','Second'],columns=['Name','Occupation'])      #undefined column name
print(data)

```

```python
#DataFrame from Dict of Series
import pandas as pd
dict={'one':pd.Series([1,2,3,4],index=['a','b','c','d']),
     'two':pd.Series([5,6,7],index=['a','b','c'])}
data=pd.DataFrame(dict)
print(data)

print('Printing based on key value in dict')
print(data['one'])

data['three']=pd.Series([1,2,3,4],index=['a','b','c','d'])   #index should be provided
print('After adding third column')
print(data)

data['four']=data['one']+data['three']
print('After adding fourth column')
print(data)

#Deleting a column using del function and pop function
del data['one']
print('After deleting column one')
print(data)

data.pop('two')
print('After deleting column two')
print(data)

#Selecting rows
print(data.loc['c'])    #location based on index value
print(data.iloc[0])     #location based on integer...1st row with integer value=0

print(data[2:4])    #Slicing rows
```

```python
#Basic Functionality-Series
import pandas as pd
import numpy as np
s=pd.Series(np.random.randn(6))
print(s)

#list of the labels of the series
print ("The axes are:")
print(s.axes)

#check if the object is empty or not
print ("Is the Object empty?")
print(s.empty)

#number of dimensions of the object..for series it is 1
print('No. of dimensions of the object: %d' %s.ndim)     #or print(s.ndim)

#size of the series
print('Size of the Series is: %d' %s.size) 

#Series as array
print('The series is: %s' %s.values) 

#first n rows
print('First 2 rows are: %s' %s.head(2)) 
print(s.head())     #by default it prints first 5 rows

#last n rows
print('Last 3 rows are: %s' %s.tail(3)) 
print(s.tail())     #by default it prints last 5 rows

```

```python
#Basic Functionality-DataFrame
import pandas as pd
dict={'Name':['Aa','Bb','Cc'],
     'Age':[10,20,30],
     'Gender':['M','F','F']}
data=pd.DataFrame(dict)
print('Original DataFrame is:')
print(data)

#Transpose
print('Transposed DataFrame is:')
print(data.T)

#Axes labels
print ("Row axis labels and column axis labels are:")
print(data.axes)

#data types
print ("Data types are:")
print(data.dtypes)

#check if the object is empty or not
print ("Is the Object empty?")
print(data.empty)

#number of dimensions of the object..for dataframe it is 2
print('No. of dimensions of the object: %d' %data.ndim)     #or print(data.ndim)

#size of the dataframe
print('Size of the DataFrame is: %d' %data.size) 

#shape of the dataframe
print ("The shape of the object is:")
print(data.shape)

#DataFrame as array
print('The dataFrame is:') 
print(data.values)

#first n rows
print('First 2 rows are:') 
print(data.head(2))
print(data.head())     #by default it prints first 5 rows

#last n rows
print('Last 1 row is:') 
print(data.tail(1))
print(data.tail())     #by default it prints last 5 rows
```

```python
#Descriptive Statistics
import pandas as pd
dict={'Name':['Aa','Bb','Cc','Dd','Ee','Ff','Gg','Hh'],
     'Age':[10,20,30,40,50,60,70,80],
     'Gender':['M','F','F','M','M','F','F','M'],
     'Marks':[20.00,25.10,30.20,35.30,40.40,45.50,50.60,55.70]}
data=pd.DataFrame(dict)
print('Original DataFrame is:')
print(data)

#Sum of the values for the requested axis
print('Sum at 0 axis is:')
print(data.sum())    #by default it takes 0 axis

print('Sum at 1 axis is:')
print(data.sum(1)) 

#Average value
print('Average is:')
print(data.mean())

#Bressel standard deviation of the numerical columns
print('Standard Deviation is:')
print(data.std())

#Number of non-null observations
print('Count is:')
print(data.count())

#Median of Values
print('Median is:')
print(data.median())

#Mode of values
print('Mode is:')
print(data.mode())

#Minimum Value
print('Minimum is:')
print(data.min())

#Maximum Value
print('Maximum is:')
print(data.max())

#Product of Values
print('Product is:')
print(data.prod())

#Cumulative Sum
print('Cumulative Sum is:')
print(data.cumsum())

#abs() and cumprod() raises exception for string or character values

#Absolute Value
#print('Absolute is:')
#print(data.abs())

#Cumulative Product
#print('Cumulative Product is:')
#print(data.cumprod())


#Summarizing Data
print('Data Summary:')
print(data.describe())    #it excludes character columns..by default ...include='number'

#use 'include' argument to include-'number','object' or 'all'
print(data.describe(include='object')) 
print(data.describe(include='all')) 
```

```python
#Function Application

import pandas as pd
import numpy as np
def adder(n1,n2):
    return n1+n2

data=pd.DataFrame(np.random.randn(5,3), columns=['col1','col2','col3'])
print(data)
print('After adding 3 to the dataframe elements:')
print(data.pipe(adder,3))

print('Mean of columns:')
print (data.apply(np.mean))  #Mean of columns
print('Mean of rows:')
print (data.apply(np.mean,axis=1))  #Mean of rows


print('Function on columns:')
print(data.apply(lambda x: x.max() - x.min()))   #Function on columns
print('Function on rows:')
print(data.apply((lambda x: x.max() - x.min()),axis=1))    #Function on rows

print(data['col1'].map(lambda x:x*100))      #Function on col1

print(data.applymap(lambda x:x*100))      #Function on complete dataframe elements
```

```python
#Reindexing
import pandas as pd
import numpy as np

N=20

data = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),  #generates date_range with starting date,no.of repetations..i.e. periods and freq of 1 Day(D)
   'x': np.linspace(0,stop=N-1,num=N),  #0 is start value,N-1 is stop value and N is no. of samples to generate
   'y': np.random.rand(N),  #generates random no for N times b/w 0 and 1
   'C': np.random.choice(['Low','Medium','High'],N).tolist(), #from choices in array/list choose N values randomly
   'D': np.random.normal(100, 10, size=(N)).tolist()   #location=100, size=10 i.e. std. deviation and o/p size=N for normal(gaussian distribution)
})

#reindex the DataFrame
data_reindexed = data.reindex(index=[0,2,5], columns=['A', 'C', 'B'])

print('Original DataFrame is:')
print(data)
print('Reindexed DataFrame is:')
print(data_reindexed)


```

```python
#Reindex to align with other object
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df2 = pd.DataFrame(np.random.randn(2,3),columns=['col1','col2','col3'])
print('df1 is:')
print(df1)
print('df2 is:')
print(df2)
df1 = df1.reindex_like(df2)
print('reindexed df1 is:')
print(df1)
```

```python
#Filling while Reindexing

import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.randn(2,3),columns=['col1','col2','col3'])
df2 = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print('df1 is:')
print(df1)
print('df2 is:')
print(df2)
print('default reindexed df1 is:')
print(df1.reindex_like(df2)) #default filling adds NaN
print('Forward Filled reindexed df1 is:')
print(df1.reindex_like(df2,method='ffill'))  #filling method=forward fill...fills the last value for next NaN values
#pad/ffill − Fill values forward

#bfill/backfill − Fill values backward

#nearest

print('Forward Filled with Limit value reindexed df1 is:')
print(df1.reindex_like(df2,method='ffill',limit=1)) #Limit specifies the maximum count of consecutive matches.

```

```python
#Iterating

import pandas as pd
import numpy as np
 
N=6
df = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),
   'C': np.random.choice(['Low','Medium','High'],N).tolist(),
   'D': np.random.normal(100, 10, size=(N)).tolist()
   })

print('Columns in dataframe:')
for col in df:            #returns cols in dataframe
   print (col)

#iterate over rows

print('Rows as key-value pairs in dataframe:')
for key,value in df.iteritems():  #iteritems() − to iterate over the (key,value) pairs
   print(key,value)
print('Rows as index-row pairs in dataframe:')
for row_index,row in df.iterrows():  #iterrows() − iterate over the rows as (index,series) pairs
   print(row_index,row)
print('Rows in dataframe:')
for row in df.itertuples():   #itertuples() − iterate over the rows as namedtuples
    print(row)
```

```python
#Sorting
import pandas as pd
import numpy as np

unsorted_df=pd.DataFrame(np.random.randn(6,2),index=[1,4,2,3,5,0],columns=['col2','col1'])
print('Unsorted dataframe is:')
print(unsorted_df)

sorted_df=unsorted_df.sort_index()   #sort rows based on index
print('Sorted dataframe is:')
print(sorted_df)

print('Sorted dataframe in descending order is:')
print(unsorted_df.sort_index(ascending=False))     #sort rows based on index in descending order

print('Sorted dataframe based on columns is:')
print(unsorted_df.sort_index(axis=1))   #sort the columns

print('Sorted dataframe based on specific column is:')
print(unsorted_df.sort_values(by='col1'))   #sort by a column

print('Sorted dataframe based on specific columns is:')
print(unsorted_df.sort_values(by=['col1','col2']))

print('Sorted dataframe is:')
print(unsorted_df.sort_values(by='col1' ,kind='mergesort')) # can choose mergesort, heapsort or quicksort 
```

```python
#Text Data

import pandas as pd
import numpy as np

s = pd.Series(['Tom', 'William Rick', 'John', 'Alber@t', np.nan, '1234','SteveSmith'])
print('Series is:')
print(s)
print('Lower Case is:')
print(s.str.lower())   #Converts strings in the Series/Index to lower case.
print('Upper Case is:')
print(s.str.upper())   #Converts strings in the Series/Index to upper case.
print('Length of the string is:')
print(s.str.len())    #Computes String length().
print('After stripping whitespaces, the string is:')
print(s.str.strip()) #Helps strip whitespace(including newline) from each string in the Series/index from both the sides.
print('After splitting string with whitespaces, the string is:')
print(s.str.split(' '))  #Splits each string with the given pattern.
print('After concatenating string with _, the string is:')
print(s.str.cat(sep='_')) #Concatenates the series/index elements with given separator.
print('One-Hot Encoding the string is:')
print(s.str.get_dummies())  #Returns the DataFrame with One-Hot Encoded values.
print('The boolean value checking the presence of whitespaces in the string is:')
print(s.str.contains(''))  #Returns a Boolean value True for each element if the substring contains in the element, else False.
print('After replacing @ with $, the string is:')
print(s.str.replace('@','$'))   #Replaces the value a with the value b.
print('After repeating each string twice, the string is:')
print(s.str.repeat(2))  #Repeats each element with specified number of times.
print('No. of appearences of m is string is:')
print(s.str.count('m'))   #Returns count of appearance of pattern in each element.
print('The boolean value checking if the string starts with T:')
print(s.str.startswith('T'))   #Returns true if the element in the Series/Index starts with the pattern.
print('The boolean value checking if the string ends with t:')
print(s.str.endswith('t'))  #Returns true if the element in the Series/Index ends with the pattern.
print('The first position of the first occurrence of the e is:')
print(s.str.find('e'))   #Returns the first position of the first occurrence of the pattern.
print('List of all occurrence of the e is:')
print(s.str.findall('e'))  #Returns a list of all occurrence of the pattern.
print('After swaping uppercase ans lowercase, the string is:')
print(s.str.swapcase)   #Swaps the case lower/upper.
print('The boolean value checking whether all characters in the string are lowercase:')
print(s.str.islower())   #Checks whether all characters in each string in the Series/Index in lower case or not. Returns Boolean
print('The boolean value checking whether all characters in the string are uppercase:')
print(s.str.isupper())   #Checks whether all characters in each string in the Series/Index in upper case or not. Returns Boolean.
print('The boolean value checking whether all characters in the string are numeric:')
print(s.str.isnumeric())   #Checks whether all characters in each string in the Series/Index are numeric. Returns Boolean.

```
